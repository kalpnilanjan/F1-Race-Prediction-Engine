```python
import pandas as pd
```

```python
df2 = pd.read_csv('pitStops.csv', header = [0],encoding = 'ISO-8859-1')
df2.head()
```

```python
# REFERENCES: https://stackoverflow.com/questions/48772271/merging-rows-and-summing-values-by-date-in-python
df2['stop'] = 1
df2.head()
```

```python
df2 = df2.groupby(['raceId','driverId']).agg({'stop': 'sum', 'milliseconds':'sum'})
df2.head()
```

```python
df2 ['milliseconds'] = df2['milliseconds']/1000
df2.head()
```

```python
df2.rename(columns={'stop':'totalStops', 'milliseconds':'pitStopTime'}, inplace=True)
df2.head()
```

```python
driverStandings = pd.read_csv('driverStandings.csv', encoding = 'ISO-8859-1')
driverStandings.head()
```

```python
df3 = pd.merge(df2, driverStandings, on='driverId')
df3.head()
```
